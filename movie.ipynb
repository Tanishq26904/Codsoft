```python
import pandas as pd 
import numpy as np
import regex as re
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
```

```python
train_data = pd.read_csv(r'D:\codesoft\movie\Genre Classification Dataset\train_data.txt', sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'], engine='python')
print(display(train_data.head()))

test_data = pd.read_csv(r'D:\codesoft\movie\Genre Classification Dataset\test_data.txt',sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'], engine='python')
print(display(test_data.head()))

test_solution_data = pd.read_csv(r'D:\codesoft\movie\Genre Classification Dataset\test_data_solution.txt',sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'], engine='python')
print(display(test_solution_data.head()))
```

Cleaning of Data

```python
train_data['DESCRIPTION'].fillna("", inplace=True)
test_data['DESCRIPTION'].fillna("", inplace=True)
```

MODEL TRAINING 

```python
tfid = TfidfVectorizer(stop_words='english', max_features=100000)
x_train = tfid.fit_transform(train_data['DESCRIPTION'])
x_test = tfid.transform(test_data['DESCRIPTION'])

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_data['GENRE'])
y_test = label_encoder.transform(test_solution_data['GENRE'])
```

```python
Mnb_classifier = MultinomialNB()
Mnb_classifier.fit(x_train, y_train)
```

```python
Mnb_classifier.predict(x_test)
```

```python
lr_classifier = LogisticRegression(max_iter=500)
lr_classifier.fit(x_train, y_train)
LogisticRegression(max_iter=500)
```

```python
lr_classifier.predict(x_test)
```

```python
def movie_pred(description):
    t_v1 = tfid.transform([description])
    pred_label = lr_classifier.predict(t_v1)
    return label_encoder.inverse_transform(pred_label)[0]

plot = "A movie about Pablo Escobar typically follows the life and rise of the infamous Colombian drug lord, who led the Medellín Cartel during the height of the cocaine trade in the 1980s and early 1990s"
print(movie_pred(plot))

plot = "where a detective must confront her secret while hunting a cunning killer who knows more about her than she realizes"
print(movie_pred(plot))
```
